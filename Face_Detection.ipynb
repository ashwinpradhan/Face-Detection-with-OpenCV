# Face Detection

```python
import cv2
import matplotlib.pyplot as plt
```

```python
#loading classifier for frontal face.
trained_faces = cv2.CascadeClassifier('haar-cascade-files-master/haarcascade_frontalface_default.xml')
```

```python
#function to convert BGR to RGB, since cv2 loads an image in BGR form.
def bgr_to_rgb(image) :
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
#loading the image to detect the face
img = cv2.imread('images/face1.png')
plt.axis('off')
plt.imshow(bgr_to_rgb(img))
plt.show()
```

```python
#converting the image into gray scaled image, since cv2 takes gray scaled image as an input.
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
#detecting the face
rect_coordinates = trained_faces.detectMultiScale(gray_image)
print(f"Face found = {len(rect_coordinates)}")

for (x, y, w, h) in rect_coordinates :
    cv2.rectangle(img, (x, y), (w+x, h+y), (0, 255, 0), 5)
plt.imshow(bgr_to_rgb(img))
plt.axis('off')
plt.show()
```

```python
#summing up this all in one function
def detect_face(image_path) :
    
    trained_faces = cv2.CascadeClassifier('haar-cascade-files-master/haarcascade_frontalface_default.xml')
    image = cv2.imread(image_path)
    image_copy = image
    
    gray_image = cv2.cvtColor(image_copy, cv2.COLOR_BGR2GRAY)
    rectangle_coordinates = trained_faces.detectMultiScale(gray_image, scaleFactor=1.5, minNeighbors=5)
    
    for (x, y, w, h) in rectangle_coordinates :
        cv2.rectangle(image_copy, (x, y), (x+w, y+h), (0, 255, 0), 2)
    
    print(f'Face detected = {len(rectangle_coordinates)}')
    rgb_image = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
    
    plt.imshow(rgb_image)
    plt.axis('off')
    plt.show()
```

```python
#testing the function by feeding image
detect_face('images/face2.jpg')
```
